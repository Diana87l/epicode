```python
import random

x=0
y=0
count=0
while count<100:
    x_in = random.randint(-10,10)
    y_in = random.randint(-10,10)
    x = x + x_in
     y= y + y_in
    count += 1
    print("ti stai muovendo in direzione:", x_in, y_in, "\n","iterazione numero:",count, "\t", "Le tue coordinate attuali sono:",x,y)
    if count == 100:
        break
print("HAI FINITO!")
```
